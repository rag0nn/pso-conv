```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
excel_file_path = "./results/result_history.xlsx"
```

```python
df = pd.read_excel(excel_file_path, index_col=None, header=None) 
df
```

```python
version= df[0]
epochs= df[1]
c1 = df[2]
c2= df[3]
c3= df[4]
test_acc= df[5]
train_acc= df[6]
f1_score= df[7]
test_loss= df[8]
train_loss=df[9]
```

```python
total_params = c1 + c2 + c3
```

```python
## TRAİN ACCURACY
plt.style.use('_mpl-gallery')

fig, ax1 = plt.subplots(1, 3, figsize=(18, 6))

ax1[0].stem(train_acc, total_params)
ax1[0].set_xlabel("Train Accuracy")
ax1[0].set_ylabel("Total Params")

# Plot for Test Accuracy
ax1[1].stem(test_acc, total_params, linefmt='g-')
ax1[1].set_xlabel("Test Accuracy")
ax1[1].set_ylabel("Total Params")

# Plot for F1 Score
ax1[2].stem(f1_score, total_params, linefmt='r-')
ax1[2].set_xlabel("F1 Score")
ax1[2].set_ylabel("Total Params")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have imported your data (train_acc, test_acc, f1_score, total_params)

# Create a unique marker style for each total_params value
markers = ['o','s','^','p','*','x','d','h','1','2','3','4','',]

# Plot for Train Accuracy
plt.style.use('_mpl-gallery')
fig, ax1 = plt.subplots(1, 4, figsize=(12, 3))

unique_total_params = list(set(total_params))  # Get unique total_params values

for i, total_param_value in enumerate(unique_total_params):
    mask = total_params == total_param_value
    ax1[0].stem(train_acc[mask], total_params[mask], markerfmt=markers[i])

ax1[0].set_xlabel("Eğitim Doğruluğu (Accuracy)")
ax1[0].set_ylabel("Toplam Parametre")


# Plot for Test Accuracy
for i, total_param_value in enumerate(unique_total_params):
    mask = total_params == total_param_value
    ax1[1].stem(test_acc[mask], total_params[mask], markerfmt=markers[i],linefmt='g-')

ax1[1].set_xlabel("Test Doğruluğu (Accuracy)")
ax1[1].set_ylabel("Toplam Parametre")


# Plot for F1 Score
for i, total_param_value in enumerate(unique_total_params):
    mask = total_params == total_param_value
    ax1[2].stem(f1_score[mask], total_params[mask], markerfmt=markers[i],linefmt='r-')

ax1[2].set_xlabel("F1 Skoru")
ax1[2].set_ylabel("Toplam Parametre")
plt.show()

# Plot for F1 Score
for i, total_param_value in enumerate(unique_total_params):
    mask = total_params == total_param_value
    ax1[3].stem(f1_score[mask], total_params[mask], markerfmt=markers[i],linefmt='r-', label=f'Total Params: {total_param_value}')

ax1[3].set_xlabel("F1 Skoru")
ax1[3].set_ylabel("Toplam Parametre")
ax1[3].legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Assuming you have imported your data (train_acc, test_acc, f1_score, total_params)

# Create a unique marker style for each total_params value
markers = ['o','s','^','p','*','x','d','h','1','2','3','4','']

# Plot for Train Accuracy
plt.style.use('_mpl-gallery')
fig, axs = plt.subplots(1, 4, figsize=(12, 3))

unique_total_params = list(set(total_params))  # Get unique total_params values

for i, total_param_value in enumerate(unique_total_params):
    mask = total_params == total_param_value
    axs[0].stem(train_acc[mask], total_params[mask], markerfmt=markers[i])
axs[0].set_xlabel("Eğitim Doğruluğu (Accuracy)")
axs[0].set_ylabel("Toplam Parametre")

# Plot for Test Accuracy
for i, total_param_value in enumerate(unique_total_params):
    mask = total_params == total_param_value
    axs[1].stem(test_acc[mask], total_params[mask], markerfmt=markers[i], linefmt='g')
axs[1].set_xlabel("Test Doğruluğu (Accuracy)")
axs[1].set_ylabel("Toplam Parametre")

# Plot for F1 Score
for i, total_param_value in enumerate(unique_total_params):
    mask = total_params == total_param_value
    axs[2].stem(f1_score[mask], total_params[mask], markerfmt=markers[i], linefmt='r')
axs[2].set_xlabel("F1 Skoru")
axs[2].set_ylabel("Toplam Parametre")
axs[2].legend()

# Değerler ve semboller Plot
for i, total_param_value in enumerate(unique_total_params):
    mask = total_params == total_param_value
    axs[3].stem(0, 0, markerfmt=markers[i],label=f'parametre sayısı: {total_param_value}')
axs[3].legend()

##çizgileri kaldır
axs[3].spines['right'].set_visible(False)
axs[3].spines['top'].set_visible(False)
axs[3].spines['left'].set_visible(False)
axs[3].spines['bottom'].set_visible(False)

#grid çizgileri kaldır
axs[3].xaxis.grid(False)
axs[3].yaxis.grid(False)

#eksen değerleri kaldır
axs[3].set_xticks([])
axs[3].set_yticks([])

plt.savefig("plot_2d_scored.jpg")


plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

### Eğitim Accuracy

```python

x = c1
y = c2
z = c3
z2=np.ones(shape=x.shape)*min(z)  ##çizgiler için

renk_ozellik = train_acc

# Renk haritası (https://matplotlib.org/stable/tutorials/colors/colormaps.html)
renk_haritasi = 'plasma'
    
# 3D scatter plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=renk_ozellik, cmap=renk_haritasi,s=50)

#çizgiler
for i,j,k,h in zip(x,y,z,z2):
    ax.plot([i,i],[j,j],[k,h], color='blue',alpha=0.5)

# Renk çubuğunu ekle
cbar = plt.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Eğitim Accuracy (Accuracy)')


    
# Eksen etiketlerini ekle
ax.set_xlabel('conv1_p')
ax.set_ylabel('conv2_p')
ax.set_zlabel('conv3_p')

plt.savefig("plot_3d_egitim_acc.jpg")
plt.show()
```
